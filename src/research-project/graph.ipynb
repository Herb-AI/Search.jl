```julia
function get_list(id)

    list = []


    for i in 1:10
        csv_file = "a_star_results/$(id)_5_100_($(i)).csv"
        df = CSV.read(csv_file, DataFrame)
        push!(list, df)
    end

    return list
end
```

```julia
using Statistics, Colors, CSV


function create_graph(id)


    dfs = get_list(id)
    dfs = [select(df, ["accuracy", "time", "cost"]) for df in dfs ]

    # Get  the mean and variance
    data_stack = cat([Array(df) for df in dfs]..., dims=3)
    mean_array = mean(data_stack, dims=3)
    variance_array = var(data_stack, dims=3)

    mean_matrix = reshape(mean_array, 100, 3)
    variance_matrix = reshape(variance_array, 100, 3)

    df = DataFrame(mean_accuracy = mean_matrix[:, 1], mean_time = mean_matrix[:, 2], mean_cost = mean_matrix[:, 3], variance_accuracy = variance_matrix[:, 1], variance_time = variance_matrix[:, 2], variance_cost = variance_matrix[:, 3])
    df = filter(row -> row[:mean_cost] < 2 && row[:mean_accuracy] > 0, df)

    # Calculate the standard deviation from the variance for accuracy, time, and cost
    df[:, :std_accuracy] = sqrt.(df[:, :variance_accuracy])
    df[:, :std_time] = sqrt.(df[:, :variance_time])
    df[:, :std_cost] = sqrt.(df[:, :variance_cost])




    # Calculate 1 - accuracy and its error band
    df[:, :mean_inaccuracy] = 1 .- df[:, :mean_accuracy]
    df[:, :ymin_inacc] = df[:, :mean_inaccuracy] .- df[:, :std_accuracy]
    df[:, :ymax_inacc] = df[:, :mean_inaccuracy] .+ df[:, :std_accuracy]

    df[!, :abs_diff_inaccuracy_cost] = abs.(df[:, :mean_inaccuracy] .- df[:, :mean_cost])
    df[!, :ymin_inaccuracy_diff] = df[:, :abs_diff_inaccuracy_cost] .- sqrt.(df[:, :variance_accuracy] .+ df[:, :variance_cost])
    df[!, :ymax_inaccuracy_diff] = df[:, :abs_diff_inaccuracy_cost] .+ sqrt.(df[:, :variance_accuracy] .+ df[:, :variance_cost])


    # Add ymin and ymax for the ribbon for accuracy, time, and cost
    df[:, :ymin_acc] = df[:, :mean_accuracy] .- df[:, :std_accuracy]
    df[:, :ymax_acc] = df[:, :mean_accuracy] .+ df[:, :std_accuracy]

    df[:, :ymin_time] = df[:, :mean_time] .- df[:, :std_time]
    df[:, :ymax_time] = df[:, :mean_time] .+ df[:, :std_time]

    df[:, :ymin_cost] = df[:, :mean_cost] .- df[:, :std_cost]
    df[:, :ymax_cost] = df[:, :mean_cost] .+ df[:, :std_cost]

    # Add an index column
    df[:, :index] = 1:nrow(df)


    df[!, :time_transformed] = 0.001 .* (1 .+ log10.(df[:, :mean_time]))
    df[!, :variance_time_transformed] = (0.001 ./ (log(10) .* df[:, :mean_time])) .^ 2 .* df[:, :variance_time]

    df[!, :ymin_time_transformed] = df[:, :time_transformed] .- sqrt.(df[:, :variance_time_transformed])
    df[!, :ymax_time_transformed] = df[:, :time_transformed] .+ sqrt.(df[:, :variance_time_transformed])


    set_default_plot_size(15cm, 5cm)

    # Create the plot
    p = plot(df, 
    # layer(x=:index, y=:mean_inaccuracy, ymin=:ymin_inacc, ymax=:ymax_inacc, 
    #     Geom.line, Geom.ribbon, 
    #     Theme(default_color=RGB(0.1, 0.8, 0.1))),
    layer(x=:index, y=:time_transformed, ymin=:ymin_time_transformed, ymax=:ymax_time_transformed, 
        Geom.line, Geom.ribbon, 
        Theme(default_color=colorant"red")),
    # layer(x=:index, y=:mean_cost, ymin=:ymin_cost, ymax=:ymax_cost, 
    #     Geom.line, Geom.ribbon, 
    #     Theme(default_color=RGB(0.1, 0.8, 0.1))),
    Guide.xlabel("Index"), 
    Guide.ylabel("Time", orientation=:vertical),
    )

    # Display the plot
    display(p)



end
create_graph(1478)
# 1499, 1510, 1478
```

```julia
using DataFrames, Statistics

# Assuming df is your DataFrame
df = DataFrame()  # your DataFrame

# Initialize columns for running mean and variance
df[!, "mean_accuracy"] = 0.0
df[!, "mean_cost"] = 0.0
df[!, "mean_time"] = 0.0
df[!, "var_accuracy"] = 0.0
df[!, "var_cost"] = 0.0
df[!, "var_time"] = 0.0

# Iterate over rows
for i in 1:nrow(df)
    df[i, "mean_accuracy"] = mean(df[1:i, "accuracy"])
    df[i, "mean_cost"] = mean(df[1:i, "cost"])
    df[i, "mean_time"] = mean(df[1:i, "time"])
    df[i, "var_accuracy"] = var(df[1:i, "accuracy"])
    df[i, "var_cost"] = var(df[1:i, "cost"])
    df[i, "var_time"] = var(df[1:i, "time"])
end
```
